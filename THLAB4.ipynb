```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import morphology
from PIL import Image

#Câu 1:Tịnh tiến và Otsu Threshold
img1 = cv2.imread("exercise/dalat.jpg")
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)


x, y, w, h = 300, 50, 200, 100
roi1 = gray1[y:y+h, x:x+w]

translated = np.zeros_like(gray1)
translated[y:y+h, x+100:x+100+w] = roi1
_, otsu = cv2.threshold(translated, int(0.3*255), 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

cv2.imwrite("lang_biang.jpg", otsu)
plt.imshow(otsu, cmap='gray')
plt.title("LangBiang - Otsu Threshold")
plt.axis('off')
plt.show()

#Câu 2:Xoay và Adaptive Thresholding
img2 = cv2.imread("exercise/dalat.jpg")
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

x2, y2, w2, h2 = 200, 200, 180, 80
roi2 = gray2[y2:y2+h2, x2:x2+w2]

center = (roi2.shape[1]//2, roi2.shape[0]//2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(roi2, M, (roi2.shape[1], roi2.shape[0]))

adaptive = cv2.adaptiveThreshold(rotated, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 60)

cv2.imwrite("ho_xuan_huong.jpg", adaptive)
plt.imshow(adaptive, cmap='gray')
plt.title("Hồ Xuân Hương - Adaptive")
plt.axis('off')
plt.show()

#Câu 3:Mapping + Binary Closing
img3 = cv2.imread("exercise/dalat.jpg")
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

x3, y3, w3, h3 = 400, 300, 150, 100
roi3 = gray3[y3:y3+h3, x3:x3+w3]

roi3_zoom = cv2.resize(roi3, (roi3.shape[1]*2, roi3.shape[0]*2), interpolation=cv2.INTER_LINEAR)

_, binary = cv2.threshold(roi3_zoom, 127, 255, cv2.THRESH_BINARY)
closed = morphology.binary_closing(binary / 255.0, morphology.square(3)) # type: ignore
closed = (closed * 255).astype(np.uint8)

cv2.imwrite("quan_truong_lam_vien.jpg", closed)
plt.imshow(closed, cmap='gray')
plt.title("Quảng Trường Lâm Viên - Closing")
plt.axis('off')
plt.show()
```
